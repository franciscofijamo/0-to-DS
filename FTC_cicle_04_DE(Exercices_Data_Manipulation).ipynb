# Data amanipulation

```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
#FILTERS Separating steps
conditions_true = (data['Delivery_person_Age'] > '35') &( data['Road_traffic_density'] == 'High ')
cols_return = ['ID', 'Delivery_person_Age']


data.loc[conditions_true, cols_return]
```

```python
# FILTERS in one line
data.loc[(data['Delivery_person_Age'] > '35') &( data['Road_traffic_density'] == 'High '), ['ID', 'Delivery_person_Age']]
```
